```python
import matplotlib.pyplot as plt
import numpy as np

# rainfall erosivity factor
rainfallR = {
    "Low": [(0, 1), (400, 1), (800, 0)],
    "Medium": [(600, 0), (1000, 1), (1400, 0)],
    "High": [(1200, 0), (1600, 1), (2000, 1)] 
}

#soil erodibility factor
soilK = {
    "Low": [(0, 1), (0.2, 1), (0.4, 0)],
    "Medium": [(0.2, 0), (0.4, 1), (0.6, 0)],
    "High": [(0.4, 0), (0.6, 1), (1.0, 1)]
}

# slope length factor
slopeLS = {
    "Low": [(0, 1), (2, 1), (4, 0)], 
    "Medium": [(2, 0), (5, 1), (8, 0)], 
    "High": [(4, 0), (8, 1), (10, 1)] 
}

# land cover factor
coverC = {
    "Good": [(0, 1), (0.1, 1), (0.5, 0)],
    "Moderate": [(0.3, 0), (0.6, 1), (0.8, 0)],
    "Poor": [(0.6, 0), (0.8, 1), (1.0, 1)] 
}

# support practice factor
supportP = {
    "Good": [(0, 1), (0.2, 1), (0.4, 0)],
    "Moderate": [(0.3, 0), (0.5, 1), (0.7, 0)],
    "Poor": [(0.6, 0), (0.8, 1), (1.0, 1)]
}

riskLevel = {
    "Low Risk": [(0, 1), (0.2, 1), (0.4, 0)],
    "Medium Risk": [(0.3, 0), (0.5, 1), (0.7, 0)],
    "High Risk": [(0.6, 0), (0.8, 1), (1.0, 1)]
}
```

```python
factors = {
    "Rainfall (R)": rainfallR,
    "Soil (K)": soilK,
    "Slope (LS)": slopeLS,
    "Cover (C)": coverC,
    "Support (P)": supportP,
    "Erosion Risk": riskLevel
}

fig, axs = plt.subplots(len(factors), 1, figsize=(10, 15)) 

for i, (title, fuzzy_sets) in enumerate(factors.items()):
    for label, points in fuzzy_sets.items():
        x, y = zip(*points)
        axs[i].plot(x, y, label=label)
    axs[i].set_title(title)
    axs[i].set_ylim(-0.1, 1.1)
    axs[i].legend()
    axs[i].grid(True)

plt.tight_layout()
plt.show()
```
